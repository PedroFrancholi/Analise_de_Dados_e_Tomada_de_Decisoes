Criamos a coluna Resultado

```python
df['Resultado'] = ''
df.loc[df['Outcome']==1,'Resultado'] = 'Diabético'
df.loc[df['Outcome']==0,'Resultado'] = 'Não Diabético'
sns.countplot(data=df, x='Resultado',palette='Blues')
```

```python
sns.scatterplot(data=df,x='Glucose',y='Insulin')
```
